```python
data_exercise_1 = data.loc[(data['AGE_NUM_'] > 30) & (data['OFFENSE_CATEGORY_NAME_'] == 'Assault Offenses'), ['ARRESTEE_ID_', 'ARREST_DATE_']]
print(data_exercise_1)
```

```python
data_exercise_2 = data.loc[(data['AGE_ID_'] == 5) & (data['MULTIPLE_INDICATOR_'] == 'N'), ['ARRESTEE_ID_', 'OFFENSE_CATEGORY_NAME_']]
print(data_exercise_2)
```

```python
data_exercise_3 = data.loc[(data['ARREST_TYPE_ID_'] == 1) | (data['OFFENSE_TYPE_ID_'] == 27), :]
print(data_exercise_3)
```

```python
data_exercise_4 = data.loc[(data['RACE_CODE_'] == 'W') & (data['OFFENSE_GROUP_'] == 'A'), ['ARRESTEE_ID_', 'RACE_DESC_', 'OFFENSE_CATEGORY_NAME_']]
print(data_exercise_4)
```

```python
data_exercise_5 = data.loc[(data['DATA_YEAR_'] == 2020) & (data['AGE_NUM_'] < 25), ['ARRESTEE_ID_', 'AGE_NUM_', 'OFFENSE_CATEGORY_NAME_']]
print(data_exercise_5)
```

```python
data_exercise_6 = data.loc[(data['OFFENSE_CATEGORY_NAME_'].isin(['Motor Vehicle Theft', 'Destruction/Damage/Vandalism of Property'])), ['ARRESTEE_ID_', 'OFFENSE_CATEGORY_NAME_']]
print(data_exercise_6)
```

```python
data_exercise_7 = data.loc[(data['MULTIPLE_INDICATOR_'] == 'N') & (data['AGE_NUM_'] != 21), ['ARRESTEE_ID_', 'MULTIPLE_INDICATOR_', 'AGE_NUM_']]
print(data_exercise_7)
```

```python
data_exercise_8 = data.loc[(data['RACE_CODE_'] != 'B') & (data['OFFENSE_GROUP_'] != 'A'), ['ARRESTEE_ID_', 'RACE_CODE_', 'OFFENSE_GROUP_']]
print(data_exercise_8)
```

```python
data_exercise_9 = data.loc[(data['AGE_ID_'] == 5) & (data['AGE_NUM_'].between(20, 30)), ['ARRESTEE_ID_', 'AGE_NUM_']]
print(data_exercise_9)
```

```python
data_exercise_10 = data.loc[(data['HC_FLAG_'] == 'Y') | (data['OFFENSE_TYPE_ID_'] < 10), ['ARRESTEE_ID_', 'HC_FLAG_', 'OFFENSE_TYPE_ID_']]
print(data_exercise_10)
```
