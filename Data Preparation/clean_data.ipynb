```python
import pandas as pd

def clean_column(data, column_name, replacements):
    print(f"Original {column_name} values:")
    print(data[column_name].value_counts())

    cleaned_data = data.copy()
    cleaned_data[column_name] = cleaned_data[column_name].replace(replacements)

    print(f"\nCleaned {column_name} values:")
    print(cleaned_data[column_name].value_counts())
    
    return cleaned_data

file_path = 'lessons_data_arrestee.csv'
data = pd.read_csv(file_path)

gender_replacements = {'Male': 'M', 'Malle': 'M', 'Maale': 'M', 'Female': 'F', 'Feemale': 'F'}
data = clean_column(data, 'SEX_CODE_', gender_replacements)


race_replacements = {'Black or African  American': 'Black or African American', 'Wh ite': 'White', 'Bla ck or African American': 'Black or African American',
                     'Black or African Amer ican': 'Black or African American', 'Whi te': 'White', 'Black or African Americ an': 'Black or African American'}
data = clean_column(data, 'RACE_DESC_', race_replacements)

arrest_type_replacements = {'Taken INTO  Custody': 'Taken INTO Custody', 'Summoned  / Cited': 'Summoned / Cited', 'On  View': 'On View', 
                             'Taken   INTO   Custody': 'Taken INTO Custody', 'Taken  INTO   Custody': 'Taken INTO Custody', 'Taken INTO    Custody': 'Taken INTO Custody',
                             'Taken   INTO Custody': 'Taken INTO Custody', 'Takenn  INTO Custody': 'Taken INTO Custody', 'On Vieww': 'On View', 'On Viiiew': 'On View',
                             'Taken INTOO Custody': 'Taken INTO Custody'}
data = clean_column(data, 'ARREST_TYPE_NAME_', arrest_type_replacements)

offense_name_replacements = {'Shoplif ting': 'Shoplifting', 'Shopli fting': 'Shoplifting', 'S imple Assault': 'Simple Assault', 'Simple As sault': 'Simple Assault',
                              'Simple Assa ult': 'Simple Assault', 'Simple Assaul t': 'Simple Assault', 'All Oth er Larceny': 'All Other Larceny',
                              'Destruction/ Damage /Va ndalism of Property': 'Destruction/Damage/Vandalism of Property', 'Destruction/ Damage/Van dalism of Property': 'Destruction/Damage/Vandalism of Property',
                              
                              }
data = clean_column(data, 'OFFENSE_NAME_', offense_name_replacements)

cleaned_file_path = 'cleaned_data.csv'
data.to_csv(cleaned_file_path, index=False)
```
